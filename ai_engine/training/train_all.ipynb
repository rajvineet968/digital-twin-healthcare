```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# ---------------------------
# Helper function
# ---------------------------
def encode_dataframe(df):
    encoders = {}
    for col in df.columns:
        if df[col].dtype == "object":
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col].astype(str))
            encoders[col] = le
    return df, encoders


# ===========================
# 1. HEART DATASET
# ===========================
heart = pd.read_csv("../data/heart.csv")

print("Heart columns:", heart.columns)

# Drop useless columns safely
heart = heart.drop(columns=[c for c in ["id", "dataset"] if c in heart.columns], errors="ignore")

# Encode categorical
heart, heart_encoders = encode_dataframe(heart)

# Target column detection
target_heart = [c for c in heart.columns if c.lower() in ["target", "output", "num"]][0]

Xh = heart.drop(target_heart, axis=1)
yh = heart[target_heart]

# Convert to binary
yh = yh.apply(lambda x: 1 if x > 0 else 0)

heart_model = RandomForestClassifier(n_estimators=100)
heart_model.fit(Xh, yh)

joblib.dump(heart_model, "../models/heart.pkl")
joblib.dump(heart_encoders, "../models/heart_encoders.pkl")

print("✅ Heart model done")


# ===========================
# 2. DIABETES DATASET
# ===========================
diabetes = pd.read_csv("../data/diabetes.csv")

print("Diabetes columns:", diabetes.columns)

target_diabetes = "Outcome"

Xd = diabetes.drop(target_diabetes, axis=1)
yd = diabetes[target_diabetes]

diabetes_model = RandomForestClassifier(n_estimators=100)
diabetes_model.fit(Xd, yd)

joblib.dump(diabetes_model, "../models/diabetes.pkl")

print("✅ Diabetes model done")


# ===========================
# 3. KIDNEY DATASET
# ===========================
kidney = pd.read_csv("../data/kidney.csv")

print("Kidney columns:", kidney.columns)

# Clean common text values
kidney = kidney.replace({
    "ckd": 1, "notckd": 0,
    "yes": 1, "no": 0,
    "present": 1, "notpresent": 0,
    "?": None
})

# Fill missing
kidney = kidney.fillna(0)

# Encode categorical
kidney, kidney_encoders = encode_dataframe(kidney)

# Target column detection
target_kidney = [c for c in kidney.columns if "class" in c.lower() or "target" in c.lower()][0]

Xk = kidney.drop(target_kidney, axis=1)
yk = kidney[target_kidney]

kidney_model = RandomForestClassifier(n_estimators=100)
kidney_model.fit(Xk, yk)

joblib.dump(kidney_model, "../models/kidney.pkl")
joblib.dump(kidney_encoders, "../models/kidney_encoders.pkl")

print("✅ Kidney model done")


print("\n🔥 ALL MODELS TRAINED SUCCESSFULLY!")
```
